```python
import numpy as np
import pandas as pd
```

```python
X = 5 * np.random.rand(30,1)
y = 4 + 1.2 * X + np.random.randn(30,1)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(7,7))

ax.set_xlim([-10,10])
ax.set_ylim([-10,10])

plt.plot(X,y,'o')
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
X[:,0]
```

```python
data = pd.DataFrame({"X" : X[:,0], "y" : y[:,0]})
```

```python
data.head()
```

```python
data.X = data.X - data.X.mean()
data.y = data.y - data.y.mean()
```

```python
fig, ax = plt.subplots(figsize=(7,7))
plt.plot(data.X,data.y,'o')
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
plt.grid(True)
plt.show()
```

```python
cov_data = data.T.dot(data) / 30 # At dot product A 공분산 행렬
"""
X : 행렬 A의 X특성 열벡터
Y : 행렬 A의 Y특성 열벡터
* : dot product

[X * X , X * Y]
[Y * X , Y * Y]
"""
cov_data
```

```python
eig = np.linalg.eig
matrix = np.matrix
covmatrix = matrix(cov_data)
```

```python
covmatrix
```

```python
value = eig(covmatrix)[0] # 고윳값
vector = eig(covmatrix)[1] # 고유벡터
```

```python
print(value)
print(vector)
```

```python
vector[0,0]
```

```python
c= 5
a = [-vector[0,0] * c,vector[0,0] * c]
b = [-vector[1,0] * c,vector[1,0] * c]

a2 = [-vector[0,1] * c,vector[0,1] * c]
b2 = [-vector[1,1] * c,vector[1,1] * c]
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
plt.plot(data.X,data.y,'o')
plt.plot(a,b)
plt.plot(a2,b2)
plt.grid(True)
```

```python
inv_vector=np.linalg.inv(vector)
```

```python
Rvector=inv_vector.dot(vector)
```

```python
Rdata = inv_vector.dot(data.T)
```

```python
a = [-Rvector[0,0] * c,Rvector[0,0] * c]
b = [-Rvector[1,0] * c,Rvector[1,0] * c]

a2 = [-Rvector[0,1] * c,Rvector[0,1] * c]
b2 = [-Rvector[1,1] * c,Rvector[1,1] * c]
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
plt.plot(Rdata[1],Rdata[0],'o')
plt.plot(a,b)
plt.plot(a2,b2)
plt.grid(True)
```

```python
x = 5 * np.random.rand(100,1)
y = 2 + 2 * x + np.random.randn(100,1)
z = 5 + x  + np.random.randn(100,1)
```

```python
data=pd.DataFrame({'x' : x[:,0],'y' : y[:,0],'z' : z[:,0]})
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([-10,10])
ax.plot(data.x, data.y, data.z,'o')
plt.show()
```

```python
data.x = data.x - data.x.mean()
data.y = data.y - data.y.mean()
data.z = data.z - data.z.mean()
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([-10,10])
plt.plot(data.x,data.y,data.z,'o')
plt.show()
```

```python
cov_data = data.T.dot(data) / 100
print(cov_data)
```

```python
eig = np.linalg.eig
matrix = np.matrix
covmatrix = matrix(cov_data)
```

```python
value = eig(covmatrix)[0] # 고윳값
vector = eig(covmatrix)[1] # 고유벡터
```

```python
print(value)
print(vector)
```

```python
c= 10
x1 = [-vector[0,0] * c,vector[0,0] * c]
y1 = [-vector[1,0] * c,vector[1,0] * c]
z1 = [-vector[2,0] * c,vector[2,0] * c]

x2 = [-vector[0,1] * c,vector[0,1] * c]
y2 = [-vector[1,1] * c,vector[1,1] * c]
z2 = [-vector[2,1] * c,vector[2,1] * c]

x3 = [-vector[0,2] * c,vector[0,2] * c]
y3 = [-vector[1,2] * c,vector[1,2] * c]
z3 = [-vector[2,2] * c,vector[2,2] * c]
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([-10,10])
plt.plot(data.x,data.y,data.z,'o')
plt.plot(x1,y1,z1)
plt.plot(x2,y2,z2)
plt.plot(x3,y3,z3)
plt.show()
```

```python
inv_vector=np.linalg.inv(vector)
Rvector=inv_vector.dot(vector)
Rdata = inv_vector.dot(data.T)
```

```python
Rdata = pd.DataFrame(Rdata.T)
Rdata.columns = ['x','y','z']
```

```python
x1 = [-Rvector[0,0] * c,Rvector[0,0] * c]
y1 = [-Rvector[1,0] * c,Rvector[1,0] * c]
z1 = [-Rvector[2,0] * c,Rvector[2,0] * c]

x2 = [-Rvector[0,1] * c,Rvector[0,1] * c]
y2 = [-Rvector[1,1] * c,Rvector[1,1] * c]
z2 = [-Rvector[2,1] * c,Rvector[2,1] * c]

x3 = [-Rvector[0,2] * c,Rvector[0,2] * c]
y3 = [-Rvector[1,2] * c,Rvector[1,2] * c]
z3 = [-Rvector[2,2] * c,Rvector[2,2] * c]
```
